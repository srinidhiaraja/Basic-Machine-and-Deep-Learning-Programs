```python
from skimage.io import imread , imshow
```

```python
tajc=imread('D:/s/mldll/ml dataset/taj.jpg ', as_gray=False)
print(tajc)
imshow(tajc)
```

```python
tajg=imread('D:/s/mldll/ml dataset/taj.jpg',as_gray=True)
print(tajg)
imshow(tajg)
```

```python
print(tajc.shape)
print(tajg.shape)
```

```python
from skimage.transform import resize,rescale
tajsz=resize(tajc,(120,120))
tajsc=rescale(tajg,scale=(0.5,0.75))
```

```python
imshow(tajc)
```

```python
imshow(tajsc)
```

```python
imshow(tajsz)
```

```python
from skimage.transform import rotate
tajr=rotate(tajc,angle=45)
imshow(tajr)
```

```python
from numpy import fliplr,flipud
tajflr=fliplr(tajc)
imshow(tajflr)
```

```python
tajfud=flipud(tajc)
imshow(tajfud)
```

```python
from skimage import exposure
tajb=exposure.adjust_gamma(tajc,gamma=0.25,gain=1)
imshow(tajb)
```

```python
from skimage import exposure
tajd=exposure.adjust_gamma(tajc,gamma=2,gain=1)
imshow(tajd)
```

```python
tajds=exposure.adjust_gamma(tajc,gamma=4,gain=1)
imshow(tajds)
```

```python
imshow(tajg)
```

```python
from skimage.filters import sobel_h,sobel_v
tajsobh=sobel_h(tajg)
imshow(tajsobh)
```

```python
tajsobv=sobel_v(tajg)
imshow(tajsobv)
```

```python
from skimage.filters import prewitt_h , prewitt_v
tajpbh=prewitt_h(tajg)
imshow(tajpbh)
```

```python
tajpv=prewitt_v(tajg)
imshow(tajpv)
```
