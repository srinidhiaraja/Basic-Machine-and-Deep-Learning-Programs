```python
import nltk
import re
```

```python
txtfile=open("D:/s/mldll/ml dataset/text mining/textminingwiki.txt")
txtdoc=txtfile.read()
print(txtdoc)
```

```python
txtdoc1=txtdoc.translate(punctuation).lower().split()
print(txtdoc1)
print(len(txtdoc1))
```

```python
word_count=Counter(txtdoc1)
print(word_count)
```

```python
txt1=re.sub(r'[^a-zA-Z0-9\s]','',txtdoc,re.I|re.A)
```

```python
print(word_count.most_common(10))
```

```python
txt1
```

```python
txt1=txt1.lower()
```

```python
txt1=txt1.strip()
```

```python
nltk.download('punkt')
doc_tokens=nltk.word_tokenize(txt1)
print(doc_tokens)
```

```python
nltk.download('averaged_perceptron_tagger')
nltk_pos_tagged=nltk.pos_tag(doc_tokens)
print(doc_tokens)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
```

```python
start_all_words=''
stopwords=set(STOPWORDS)
for word in doc_tokens:
    start_all_words=start_all_words+word+' '
start_word_cloud=WordCloud(width=1000,height=1000,background_color='white',stopwords=stopwords,min_font_size=10).generate(start_all_words)
```

```python
print(start_all_words)
```

```python
plt.imshow(start_word_cloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
nltk.download('stopwords')
```

```python
eng_s=nltk.corpus.stopwords.words('english')
final_tokens=[token for token in doc_tokens
             if token not in eng_s]
```

```python
print(final_tokens)
```

```python
newdoc=' '.join(final_tokens)
print(newdoc)
print(len(newdoc))
```

```python
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
ps=PorterStemmer()
lem=WordNetLemmatizer()
lemlist=[]
pslist=[]
```

```python
for word in final_tokens:
    lemlist.append(lem.lemmatize(word,'v'))
    pslist.append(ps.stem(word))
```

```python
print(lemlist[1:20])
print(pslist[1:20])
```
